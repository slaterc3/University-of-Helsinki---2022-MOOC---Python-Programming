```python
# Write your solution here
def get_input():
    user_nums = input("Exam points and exercises completed: ")
    exams_exercises = []
    total_scores = []
    while user_nums != '':
        nums = [int(x) for x in user_nums.split()]
        exams_exercises.append(nums)
        user_nums = input("Exam points and exercises completed: ")
    for scores in exams_exercises:
        scores[1] //= 10
        if scores[0] < 10 or (scores[0]+scores[1] < 15):
            scores.append(False)
        else:
            scores.append(True)
    for scores in exams_exercises:
        total_scores.append([scores[0] + scores[1],scores[2]])
#         scores = sum(scores)
    return total_scores
def avg_points(lst):
#     points = []
    total = 0
    for score in lst:
        total += score[0]
    return total / len(lst)

def pass_percentage(lst):
    passing = 0
    for score in lst:
        if score[1]:
            passing += 1
    return 100 * (passing / len(lst))

def get_grades(lst):
    grades = {5:0, 4:0, 3:0, 2:0, 1:0, 0:0}
    
    for scores in lst:
        if scores[1]:
            if scores[0] > 27:
                grades[5] += 1
            elif scores[0] > 23:
                grades[4] += 1
            elif scores[0] > 20:
                grades[3] += 1
            elif scores[0] > 17:
                grades[2] += 1
            elif scores[0] > 14:
                grades[1] += 1
            else:
                grades[0] += 1
        else:
            grades[0] += 1
    return grades

def main():
    x = get_input()
    avg = avg_points(x)
    passing_rate = pass_percentage(x)
    grades = get_grades(x)
    print("Statistics:")
    print(f"Points average: {avg:.1f}")
    print(f"Pass percentage: {passing_rate:.1f}")
    print("Grade distribution:")
    for k, v in grades.items():
        print(f' {k}: {v * "*"}')
main()



```
