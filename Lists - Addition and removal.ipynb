```python
# Write your solution here
lst = []
print(f"The list is now {lst}")
option = input("a(d)d, (r)emove or e(x)it: ")
i = 1
while option != 'x':
    if lst == []:
        i = 1
        if option == 'd':
            lst.insert(0, i)
            i = 2
    else:
        if option == 'd':
            lst.insert(lst[-1], lst[-1] + 1)
        elif option == 'r':
            lst.pop()
        else:
            break
    print(f"The list is now {lst}")
    option = input("a(d)d, (r)emove or e(x)it: ")
print("Bye!")
```
