```python
# Write your solution here
def longest_series_of_neighbours(lst):
    i = 1
    long = 1
    count = 1
    while i < len(lst):
        if abs(lst[i] - lst[i-1]) == 1:
            count += 1
        else:
            long = max(count, long)
            count = 1
        i += 1
    return max(count, long)
if __name__ == "__main__":
    # pass
    my_list = [1,2,5,7,6,5,6,3,4,1,0]
    print(longest_series_of_neighbours(my_list))
```
