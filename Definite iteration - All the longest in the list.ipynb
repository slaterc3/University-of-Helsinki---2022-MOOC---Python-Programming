```python
# Write your solution heredef length_of_longest(lst):
def all_the_longest(lst):
    long_lst = []
    longest = len(lst[0])

    for i in range(1, len(lst)):
        if len(lst[i]) > longest:
            longest = len(lst[i])
            # long_lst.append(lst[i])
    for string in lst:
        if len(string) == longest:
            long_lst.append(string)
    return long_lst



if __name__ == "__main__":
    # pass
    my_list = ['adele', 'mark', 'dorothy',
    'tim', 'hedy', 'richard']
    res = all_the_longest(my_list)
    print(res)
```
