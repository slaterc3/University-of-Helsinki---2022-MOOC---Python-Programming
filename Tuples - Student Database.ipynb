```python
def add_student(students:dict, student:str):
    if student not in students:
        students[student] = []
    return students
def add_course(students:dict, student:str, course_tup:tuple):
    if student in students:
        if course_tup[1] > 0:
            for course in students[student]:
                if course[0] == course_tup[0]:
                    if course[1] <= course_tup[1]:
                        
                        students[student].remove(course)
                        students[student].append(course_tup)
                        return students
                    else:
                        return
            students[student].append(course_tup)
            
    return students

def print_student(students:dict, student:str):
    if student in students:
        print(f"{student}:")
        # if students[student]:
        if len(students[student]) == 0:
            print(" no completed courses")
        elif len(students[student]) == 1:
            print(" 1 completed courses:")
            print(f"  {students[student][0][0]} {students[student][0][1]}")
#             print(f"  {students[student][0]} {students[student[1]]}")
#             print(f" average grade {students[student][1]}")
            print(f" average grade {float(students[student][0][1])}")
        else:
            grades = []
            for tup in students[student]:
                grades.append(tup[1])
            avg_grade = sum(grades) / len(students[student])
            print(f" {len(students[student])} completed courses:")
            for course in students[student]:
                print('  ' + course[0] + ' ' + str(course[1]))
            print(f" average grade {float(avg_grade)}")
        # else:
        #     print(" no completed courses")
    else:
        print(student + ': no such person in the database')
def summary(students):
    num_students = len(students)
#     avg_grade_student = []
    most_courses = 0
    best_student = ''
    for student in students:
        if len(students[student]) > most_courses:
            most_courses = len(students[student])
            best_student = student
    best_avg = 0.0
    best_avg_student = ''
#     avg_grade = 0
    for student in students:
        avg = 0
        total = 0
        for i in range(len(students[student])):
            total += students[student][i][1]
        avg = total / (len(students[student]))
        if avg > best_avg:
            best_avg = avg
            best_avg_student = student
    print(f'students {num_students}')
    print(f"most courses completed {most_courses} {best_student}")
    print(f"best average grade {best_avg} {best_avg_student}")
            
    
```

```python
# Part 1: adding students
students = {}
add_student(students, "Peter")
add_student(students, "Eliza")
print_student(students, "Peter")
print_student(students, "Eliza")
print_student(students, "Jack")
```

```python
# Part 2: adding courses
students = {}
add_student(students, "Peter")
add_course(students, "Peter", ("Introduction to Programming", 3))
add_course(students, "Peter", ("Advanced Course in Programming", 2))
print_student(students, "Peter")
```

```python
# Part 3: repeating courses
students = {}
add_student(students, "Peter")
add_course(students, "Peter", ("Introduction to Programming", 3))
add_course(students, "Peter", ("Advanced Course in Programming", 2))
add_course(students, "Peter", ("Data Structures and Algorithms", 0))
add_course(students, "Peter", ("Introduction to Programming", 2))
print_student(students, "Peter")
```

```python
# Part 4: gradebook database summary
students = {}
add_student(students, "Peter")
add_student(students, "Eliza")
add_course(students, "Peter", ("Data Structures and Algorithms", 1))
add_course(students, "Peter", ("Introduction to Programming", 1))
add_course(students, "Peter", ("Advanced Course in Programming", 1))
add_course(students, "Eliza", ("Introduction to Programming", 5))
add_course(students, "Eliza", ("Introduction to Computer Science", 4))
summary(students)
```
