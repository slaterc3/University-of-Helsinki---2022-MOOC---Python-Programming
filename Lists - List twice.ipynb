```python
# Write your solution here
items = []
new_item = int(input("New item: "))

while new_item != 0:
    items.append(new_item)
    sorted_items = sorted(items)
    print(f"The list now: {items}")
    print(f"The list in order: {sorted_items}")
    new_item = int(input("New item: "))
print('Bye!')
```
