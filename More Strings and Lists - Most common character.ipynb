```python
# Write your solution here
def most_common_character(str1):
    char_dict = {}
    # count = 0
    most = 0
    letter = ''

    for char in str1:
        if char not in char_dict: 
            char_dict[char] = 1
        else:
            char_dict[char] += 1
    for k, v in char_dict.items():
        if v >= most:
            most = v
            letter = k
    # return char_dict
    return letter
# sec = "exemplaryelementary"
# x = most_common_character(sec)
# # for k, v in x.items():
# print(x)
    # print(k,v)
# print(x)
if __name__ == "__main__":
    first_string = 'abcdeded'
    print(most_common_character(first_string))
```
